```python
import cv2
from ultralytics import YOLO

# Inisialisasi model YOLO
model = YOLO("model.pt")

# Daftar nama kelas
class_names = [
    "Larangan Berhenti",
    "Larangan Masuk Untuk Kendaraan",
    "Larangan Parkir",
    "Lampu Hijau",
    "Lampu Kuning",
    "Lampu Merah",
    "Larangan Belok Kanan",
    "Larangan Belok Kiri",
    "Larangan Berjalan Terus",
    "Larangan Memutar Balik",
    "Peringatan Alat Pemberi Syarat",
    "Peringatan Banyak Pejalan Kaki",
    "Peringatan Pintu Perlintasan Kereta Api",
    "Peringatan Simpang Tiga Sisi Kiri",
    "Peringatan Penegasan Rambu Tambahan",
    "Perintah Masuk Jalur Kiri",
    "Perintah Pilihan Memasuki Salah Satu Jalur",
    "Petunjuk Area Parkir",
    "Petunjuk Lokasi Pemberhentian Bus",
    "Petunjuk Lokasi Putar Balik",
    "Petunjuk Penyeberangan Pejalan Kaki"
]

# Buka file video atau kamera
# cap = cv2.VideoCapture("C:\\Users\\ASUS\\Videos\\2024-06-25 09-51-33.mp4")
cap = cv2.VideoCapture(0)
# Periksa apakah video berhasil dibuka
if not cap.isOpened():
    print("Error: Tidak dapat membuka video.")
    exit()

# Buat jendela untuk menampilkan hasil
cv2.namedWindow("YOLOv8 Object Detection", cv2.WINDOW_NORMAL)

frame_skip = 5  # Mengatur untuk memproses setiap frame ke-2
frame_count = 0

while cap.isOpened():
    # Baca frame dari video
    ret, frame = cap.read()
    if not ret:
        print("Video selesai atau terjadi kesalahan.")
        break

    # Skipping frames to reduce computation
    frame_count += 1
    if frame_count % frame_skip != 0:
        continue

    # Deteksi objek pada frame yang sudah diubah ukurannya
    results = model(frame)

    # Iterasi melalui setiap hasil deteksi dalam results
    for result in results:
        for det in result.boxes.data.tolist():
            x1, y1, x2, y2, conf, cls = det
            if conf > 0.5:
                # Gambar kotak pembatas pada frame asli
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                # Tambahkan label kelas pada frame asli
                cv2.putText(frame, f"{class_names[int(cls)]}", (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Tampilkan frame asli dengan hasil deteksi
    cv2.imshow("YOLOv8 Object Detection", frame)

    # Tekan 'q' untuk keluar dari loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Rilis sumber video dan tutup semua jendela OpenCV
cap.release()
cv2.destroyAllWindows()
```
